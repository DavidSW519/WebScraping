```python
class Limpieza():
    def __init__(self,dato):
        self.dato = dato
    def precio(self):
        lista = ['PRECIO',':',' ',',','MXN','$',')','(']
        y = self.dato.upper()
        for i in lista:
            y = y.replace(i,'').strip()
        return y
    def cp(self):
        y = (5-len(self.dato))*'0'+self.dato
        return y
    def auto(self):
        y = self.dato.upper()
        palabras = ['SEDAN','COUPE','CONVERTIBLE']
        for i in palabras:
            y = y.replace(i,'').strip()
        return y
    def moneda(self):
        return locale.currency(float(self.dato),grouping=True)
    
    def Normalizar(self):
        acentos_dict = {"Á":"A","É":"E","Í":"I","Ó":"O","Ú":"U",'á':'a','é':'e','í':'i','ó':'o','ú':'u'}
        y = self.dato
        for con_acento, sin_acento in acentos_dict.items():
            y = y.replace(con_acento, sin_acento)
        return y
    
class Wait_Scrap():
    def __init__(self,driver,seconds):
        self.driver = driver
        self.seconds = seconds
        self.wait = WebDriverWait(self.driver, self.seconds)
    def xpath(self,ruta):
        wait = self.wait
        element = wait.until(EC.presence_of_element_located((By.XPATH, ruta)))
        return element
    def class_name(self,ruta):
        wait = self.wait
        element = wait.until(EC.presence_of_element_located((By.CLASS_NAME, ruta)))
        return element
    def css_selector(self,ruta):
        wait = self.wait
        element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, ruta)))
        return element
        
class Find_Scrap():
    def __init__(self,driver):
        self.driver = driver
    def xpath_p(self,ruta):
        elements  = self.driver.find_elements(By.XPATH,ruta)
        return elements
    def xpath_s(self,ruta):
        element  = self.driver.find_element(By.XPATH,ruta)
        return element
    def css_p(self,ruta):
        elements  = self.driver.find_elements(By.CSS_SELECTOR,ruta)
        return elements
    def css_s(self,ruta):
        element  = self.driver.find_element(By.CSS_SELECTOR,ruta)
        return element
    def class_p(self,ruta):
        elements  = self.driver.find_elements(By.CLASS_NAME,ruta)
        return elements
    def class_s(self,ruta):
        element  = self.driver.find_element(By.CLASS_NAME,ruta)
        return element
    def id_p(self,ruta):
        elements  = self.driver.find_elements(By.ID,ruta)
        return elements
    def id_s(self,ruta):
        element  = self.driver.find_element(By.ID,ruta)
        return element
    

class Find_Wait_Scrap():
    def __init__(self,driver):
        self.driver = driver
    def xpath_p(self,ruta):
        elements = 0
        for second in range(1,60):
            time.sleep(1)
            try:
                elements = self.driver.find_elements(By.XPATH,ruta)
                break
            except:
                continue
        return elements
    def xpath_s(self,ruta):
        element = 0
        for second in range(1,60):
            time.sleep(1)
            try:
                element = self.driver.find_element(By.XPATH,ruta)
                break
            except:
                continue
        return element
    def css_p(self,ruta):
        elements = 0
        for second in range(1,60):
            time.sleep(1)
            try:
                elements = self.driver.find_elements(By.CSS_SELECTOR,ruta)
                break
            except:
                continue
        return elements
    def css_s(self,ruta):
        element = 0
        for second in range(1,60):
            time.sleep(1)
            try:
                element = self.driver.find_element(By.CSS_SELECTOR,ruta)
                break
            except:
                continue
        return element
    def class_p(self,ruta):
        elements = 0
        for second in range(1,60):
            time.sleep(1)
            try:
                elements = self.driver.find_elements(By.CLASS_NAME,ruta)
                break
            except:
                continue
        return elements
    def class_s(self,ruta):
        element = 0
        for second in range(1,60):
            time.sleep(1)
            try:
                element = self.driver.find_element(By.CLASS_NAME,ruta)
                break
            except:
                continue
        return element
    def id_p(self,ruta):
        elements = 0
        for second in range(1,60):
            time.sleep(1)
            try:
                elements = self.driver.find_elements(By.ID,ruta)
                break
            except:
                continue
        return elements
    def id_s(self,ruta):
        element = 0
        for second in range(1,60):
            time.sleep(1)
            try:
                element = self.driver.find_element(By.ID,ruta)
                break
            except:
                continue
        return element
    
def Click_Wait(element):
    for s in range(0,60):
        try:
            element.click()
            break
        except:
            time.sleep(1)


def Select_Wait(element):
    a=0
    for s in range(0,60):
        try:
            a = Select(element)
            break
        except:
            time.sleep(1)
    return a
```

```python
def formulario_contacto(driver):
    contact_dict = {'name':"Juan",'lastname':"Perez",'mother-lastname':"Gomez",
                'email':"juan.perez@gmail.com",'tel':"5512345678"}
    for key in contact_dict.keys():
            box = Find_Wait_Scrap(driver).xpath_s(f'//*[@id="{key}"]')
            box.send_keys(contact_dict[key])

def formulario_desc_auto(campos_dict, driver):
    for key in campos_dict.keys():
        Cerrar_Anuncio()
        lista_desplegable = Find_Wait_Scrap(driver).xpath_s(f'//*[@id="{key}"]')
        if key in ['color-selected','intside']:
            time.sleep(5)
            Cerrar_Anuncio()
            Select_Wait(lista_desplegable).select_by_index(campos_dict[key])
        else:
            Cerrar_Anuncio()
            Select_Wait(lista_desplegable).select_by_visible_text(campos_dict[key])
        time.sleep(2)
    

def formulario_estado(campos_dict, cp_index,estado,driver):
    if cp_index == 0: # Para el primer CP ya esta todo cargado
        estados_desplegable = Find_Wait_Scrap(driver).xpath_s('//*[@id="state"]')
        time.sleep(5)
        Cerrar_Anuncio()
        Select_Wait(estados_desplegable).select_by_visible_text(estado)
    else: # Para el resto de los CP se debe cargar el modelo, año, version y estado, etc.
        formulario_desc_auto(campos_dict, driver)
        estados_desplegable = Find_Wait_Scrap(driver).xpath_s('//*[@id="state"]')
        Cerrar_Anuncio()
        Select_Wait(estados_desplegable).select_by_visible_text(estado)

    dealer_desplegable = Find_Wait_Scrap(driver).xpath_s('//*[@id="dealer"]')
    Cerrar_Anuncio()   
    Select_Wait(dealer_desplegable).select_by_index(1)
    siguiente = Find_Wait_Scrap(driver).xpath_s('//*[@id="form"]/fieldset[1]/div[5]/button')
    Cerrar_Anuncio()
    Click_Wait(siguiente)

def acept_terms1(driver):
    term_cond = Find_Wait_Scrap(driver).xpath_s('//*[@id="terms1"]')
    if not term_cond.is_selected():
        Find_Wait_Scrap(driver).xpath_s('//*[@id="form"]/fieldset[2]/div[3]').click()
    # Boton siguiente 2
    siguiente = Find_Wait_Scrap(driver).xpath_s('//*[@id="form"]/fieldset[2]/div[4]/button[2]')
    #siguiente.click()
    Click_Wait(siguiente)
    time.sleep(2)

def formularo_coti(cp,driver):
    coti_dict = {'percentage-payment':'20% de enganche',
                  'installments':'12 meses'}
    cp_box = Find_Wait_Scrap(driver).xpath_s('//*[@id="cp"]')
    cp_box.clear()
    cp_box.send_keys(cp)


    for key in coti_dict.keys():
        lista_desplegable = Find_Wait_Scrap(driver).xpath_s(f'//*[@id="{key}"]')
        Select_Wait(lista_desplegable).select_by_visible_text(coti_dict[key])
        time.sleep(4)   
     
    for i in range(120):
        if 'Espere' in Find_Wait_Scrap(driver).xpath_s('//*[@id="resultado"]').text:
            time.sleep(1)
        else:
            break
    
    if 'Por favor' in Find_Wait_Scrap(driver).xpath_s('//*[@id="resultado"]').text:
        plan = 'No Disponible'
    else:
        time.sleep(2)
        # Plan
        try:
            plan_list = Find_Wait_Scrap(driver).xpath_s(f'//*[@id="plan-selected"]')
            plan = Select_Wait(plan_list).options[1].text
            Select_Wait(plan_list).select_by_index(1)
            time.sleep(1)
        except:
            plan = 'No Disponible'
    
    enganche = coti_dict['percentage-payment']
    meses = coti_dict['installments']
    
    return enganche,meses,plan

```

```python
try:
    listos = pd.read_csv(os.path.join(path_coti,'Cotizaciones_ACotizar_Arrizo8.csv'))
    listos.drop_duplicates(inplace= True)
except:
    pass


for entrada in range(len(acotizar)):
    Cerrar_Anuncio()
    modelo = acotizar.iloc[entrada]['Modelo']
    año = str(acotizar.iloc[entrada]['Año'])
    version = acotizar.iloc[entrada]['Version']
    cps = acotizar.iloc[entrada]['CodigoPostal']
    campos_dict = {'model-selected':modelo,'year':año,
                   'version':version,'color-selected':1,'intside':1}
    try:
        listos = listos[(listos['Modelo']==modelo) & (listos['Version']==version)]
        aux_cp = listos.pivot_table(index='CP',values='Aseguradora',aggfunc=pd.Series.nunique).reset_index()
        aux_cp = aux_cp[aux_cp['Aseguradora']>=4]

        cp_listos = aux_cp['CP'].unique()
        cp_listos = [str(i) for i in cp_listos]
        cps = [i for i in cps if i not in cp_listos]
    except:
        pass

    cps = [cp for cp in cps if cp_dict[cp]!='Hidalgo']
    cps = [cp for cp in cps if cp_dict[cp]!='Durango']
    

    print(modelo,año,version)
    
    # Llenamos el formulario de descripcion del auto
    formulario_desc_auto(campos_dict, driver)
    time.sleep(2)

    #|||Seleccionar Estado por cada CP|||
    num_cps = len(cps)
    for cp_index in range(num_cps):
        cp = cps[cp_index]
        estado = cp_dict[cp]

        if estado == 'SINESTADO':
            print(f'\tNo se encuentra el {cp} {cp_index+1}/{num_cps}')
            continue
        else:
            print(f'\tConsultando {cp} {cp_index+1}/{num_cps}')

        if cp_index == 0:
            time.sleep(1)
            formulario_estado(campos_dict,cp_index,estado,driver)
            time.sleep(2)
            formulario_contacto(driver)
        else:
            driver.quit()
            driver = webdriver.Chrome(options = CO)
            driver.get("https://www.chirey.mx/configurar")
            time.sleep(15)
            formulario_desc_auto(campos_dict, driver)
            time.sleep(1)
            formulario_estado(campos_dict,cp_index,estado,driver)
            time.sleep(2)
            formulario_contacto(driver)

        # Boton terminos y condiciones
        acept_terms1(driver)

        # Codigo Postal
        enganche, meses, plan = formularo_coti(cp,driver)
        if plan == 'No Disponible':
            print(f'\t\tNo Disponible {cp_index+1}/{num_cps}')
            continue
        else:
            pass
        seguros = Find_Wait_Scrap(driver).xpath_s(f'//*[@id="form"]/fieldset[3]/div[3]/div[2]/div/div')
        #seguros.click()
        Click_Wait(seguros)
        opciones_seguros = Find_Wait_Scrap(driver).class_p(f'insurance')

        num_opciones = len(opciones_seguros)
        for op in range(num_opciones):
            try:
                if op == 0 :
                    seguro = opciones_seguros[op]
                else:
                    driver.get("https://www.chirey.mx/configurar")
                    formulario_desc_auto(campos_dict, driver)
                    time.sleep(1)
                    formulario_estado(campos_dict,cp_index,estado,driver)
                    time.sleep(3)
                    formulario_contacto(driver)
                    time.sleep(1)
                    acept_terms1(driver)
                    time.sleep(1)
                    enganche, meses, plan = formularo_coti(cp,driver)
                    time.sleep(1)
                    seguros = Find_Wait_Scrap(driver).xpath_s(f'//*[@id="form"]/fieldset[3]/div[3]/div[2]/div/div')
                    #seguros.click()
                    Click_Wait(seguros)
                    opciones_seguros = Find_Wait_Scrap(driver).class_p(f'insurance')
                    seguro = opciones_seguros[op]
            except:
                continue

            
            time.sleep(1)
            mensualidad = seguro.text
            #seguro.click()
            Click_Wait(seguro)

            boton_aceptar_seguro = Find_Wait_Scrap(driver).xpath_s(f'//*[@id="offcanvasRight3"]/div[2]/button')
            #boton_aceptar_seguro.click()
            Click_Wait(boton_aceptar_seguro)

            seguros = Find_Wait_Scrap(driver).xpath_s(f'//*[@id="form"]/fieldset[3]/div[3]/div[2]/div/div')
            aseguradora = seguros.text

            boton_cotizar = Find_Wait_Scrap(driver).xpath_s(f'//*[@id="request-btn"]')
            #boton_cotizar.click()
            Click_Wait(boton_cotizar)


            for i in range(60):
                current_page = driver.current_url
                if current_page != 'https://www.chirey.mx/gracias/cotizacion':
                    time.sleep(1)
                else:
                    break
            time.sleep(3)
            Cerrar_Anuncio()
            boton_descarga = Find_Wait_Scrap(driver).xpath_s(f'//*[@id="gracias_msg"]/a')
            Cerrar_Anuncio()
            Click_Wait(boton_descarga)
            
            print(f'\t\t{aseguradora} {op+1}/{num_opciones}')
            importe_seguro_anual = 0
            time.sleep(3)
            for intento_descarga in range(15):
                if intento_descarga <= 13:
                    try:
                        importe_seguro_anual = leer_descarga(modelo, cp, aseguradora)
                        aux = pd.DataFrame({'Modelo':[modelo],'Año':[año],
                                    'Version':[version],'Estado':[estado],
                                    'CP':[cp],'Aseguradora':[aseguradora],
                                    'Mensualidad':[mensualidad],'Seguro_Anual':[importe_seguro_anual]})
                        cotizaciones = pd.concat([cotizaciones,aux],ignore_index=True)
                        break
                    except:
                        time.sleep(1)
                else:
                    aux = pd.DataFrame({'Modelo':[modelo],'Año':[año],
                                    'Version':[version],'Estado':[estado],
                                    'CP':[cp],'Aseguradora':[aseguradora],
                                    'Mensualidad':[mensualidad],'Seguro_Anual':[importe_seguro_anual]})
                    cotizaciones = pd.concat([cotizaciones,aux],ignore_index=True)

            time.sleep(5)

            driver.get("https://www.chirey.mx/configurar")
            time.sleep(5)
        


            #if op == 0:
            #    break
        try:
            cotizaciones = pd.concat([listos, cotizaciones],ignore_index=True)
            cotizaciones.drop_duplicates(inplace= True)
            cotizaciones.to_csv(os.path.join(path_coti,'Cotizaciones_ACotizar_Arrizo8.csv'),index=False)
        except:
            cotizaciones.to_csv(os.path.join(path_coti,'Cotizaciones_ACotizar_Arrizo8.csv'),index=False)
            cotizaciones.drop_duplicates(inplace= True)
```
